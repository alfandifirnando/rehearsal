```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from jcopml.pipeline import num_pipe, cat_pipe
from jcopml.utils import save_model, load_model
from jcopml.plot import plot_missing_value
from jcopml.feature_importance import mean_score_decrease
```

# Import Data

```python
df = pd.read_csv("data/mobile.csv")
df.head()
```

```python
plot_missing_value(df)
```

# Dataset Splitting

```python
X = df.drop(columns="price_range")
y = df.price_range

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from jcopml.tuning import grid_search_params as gsp
```

```python
X_train.head()
```

```python
X_train.columns
```

```python
 gsp.rf_params
```

```python
df.price_range.value_counts()
```

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(), ['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep', 'mobile_wt',
                             'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_w', 'talk_time']),
    ('categoric', cat_pipe(encoder='onehot'), ['blue', 'dual_sim', 'four_g', 'three_g',  'touch_screen', 'wifi']),
])

pipeline = Pipeline([
    ('prep', preprocessor),
    ('algo', RandomForestClassifier(n_jobs=-1, random_state=42))
])

model = GridSearchCV(pipeline, gsp.rf_params, cv=3, n_jobs=-1, verbose=1)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

# Feature Importance

```python
df_imp = mean_score_decrease(X_train, y_train, model, plot=True, topk=10)
```

# Randomized Search

### Karena Grid-Serch mencoba semua kombinasi yang kita minta, maka akan memakan waktu yang sangat lama. itulah mengapa GrdiSearch juga disebut sebagai Exhaust Search. tentunya akan lebih efisien kalau kita tidak mencoba semua kombinasi, tetapi memilih n_iter kombinasi secara acak. Search seperti itu disebut RandomizedSearch. Kelebihannya adalah kita tidak wajib memberikan angka pasti, tetapi boleh juga memberikan distribusi search nya. Kelemahannya tentu tidak menjamin kita mendapatkan parameter yang optimum karena hanya mencari di sebagian kombinasi saja.

```python
from sklearn.model_selection import RandomizedSearchCV
from jcopml.tuning import random_search_params as rsp
```

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(poly=2), ['battery_power', 'px_height', 'px_width', 'ram']),
])

pipeline = Pipeline([
    ('prep', preprocessor),
    ('algo', RandomForestClassifier(n_jobs=-1, random_state=42))
])

model = RandomizedSearchCV(pipeline, rsp.rf_poly_params, cv=3, n_iter=50, n_jobs=-1, verbose=1, random_state=42)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

# Save Model

```python
save_model(model.best_estimator_, "rf_mobile.pkl")
```
